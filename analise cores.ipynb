# Análize de cores

Pegamos várias imagens dos círculos ciano e magenta em diferentes iluminações e posições para obtermos melhores resultados

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import time as t
import sys
import math
print ("OpenCV Version : %s " % cv2.__version__)

from ipywidgets import widgets, interact, interactive, FloatSlider, IntSlider

import importlib
```

```python
def limites(img_x):
    hist,bins = np.histogram(img_x.flatten(),256,[0,256])
    
    grandes = []
    for i in range(len(hist)):
        h = hist[i]
        if h > np.mean(hist):
            grandes.append(i)

    minn = min(grandes)
    maxx = max(grandes)
    
    print(grandes)
    return [minn, maxx]

def ac_x(img, i, j):
    print(i)
    img_x = img[:,:,j]
    if j == 0:
        x = 'H'
    elif j == 1:
        x = 'S'
    elif j == 2:
        x = 'V'
   
    minh = np.min(img_x)
    maxh = np.max(img_x)
    [mh, Mh] = limites(img_x)
    print('Min '+ x +': '+ str(minh) + ' | Max '+x+': ' +str(maxh)+ '    || Maioria: '+ str(mh) + ' < '+x+' < ' + str(Mh))
    print()
    

def ac(img, i):
    img_h = img[:,:,0]
    img_s = img[:,:,1]
    img_v = img[:,:,2]
             
    # Não mostramos 0 nem 255
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(9, 3))
    axes[0].set_title(str(i)+ ' H')
    axes[1].set_title(str(i)+ ' S')
    axes[2].set_title(str(i)+ ' V')
    axes[0].hist(img_h.flatten(),253,[1,254], color = 'b')
    axes[1].hist(img_s.flatten(),253,[1,254], color = 'g')
    axes[2].hist(img_v.flatten(),253,[1,254], color = 'c')
    fig.tight_layout()
```

# Ciano

#### Carrega as imagens

```python
imgs = []
for i in range(13):
    arq = "samples/c"+str(i)+".png"
    img_bgr = cv2.imread(arq)
    img  = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
    imgs.append(img)
```

### Histograma de cada canal

```python
for i in range( len(imgs) ):
    ac(imgs[i], i)
```

#### Análise dos valores mais comuns do histograma

```python
# H
for i in range(len(imgs)):
    ac_x(imgs[i] , i ,0)
```

```python
# S
for i in range(len(imgs)):
    ac_x(imgs[i] , i ,1)
```

```python
# V
for i in range(len(imgs)):
    ac_x(imgs[i] , i , 2)
```
